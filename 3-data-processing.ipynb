* Use a WEB request to acquire the data (like for lab 1), the API endpoint is https://archive-api.open-meteo.com/v1/archive

```python
from typing import List
import requests
import json


# create a function
def acquire_data(latitude: float, longitude: float, start_date: str, end_date: str, features: List[str]) -> dict:
    """
    Acquires data from the Open-meteo API for a given location and time period.
    The API endpoint is https://archive-api.open-meteo.com/v1/archive.

    Args:
        latitude (float): latitude of the location
        longitude (float): longitude of the location
        start_date (str): start date of the data in the format YYYY-MM-DD
        end_date (str): end date of the data in the format YYYY-MM-DD
        features (List[str]): list of features to be extracted from the API
    Returns:
        (dict): dictionary of data
    """   
    # add your code here
    features = ",".join(features)
    baseurl = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude":latitude,
        "longitude": longitude,
        "start_date":start_date,
        "end_date": end_date,
        "daily": features,
        "timezone":"GMT"
    }

    weather = requests.get(baseurl,params).json()

    return weather
```

```python
from numpy.typing import ArrayLike
import matplotlib.pyplot as plt


def lineplot(x: ArrayLike, y: ArrayLike, title: str, x_label: str, y_label: str,plt_color : str) -> (plt.Figure, plt.Axes):
    """
    Creates a line plot for the given data, title and labels.

    Args:
        x (ArrayLike): value of the data on the x-axis like np.array, list, etc
        y (ArrayLike): value of the data on the y-axis
        title (str): title of the plot
        x_label (str): x-axis label
        y_label (str): y-axis label
    Returns:
        (plt.Figure, plt.Axes): matplotlib figure and axes objects
    """
    # creating the plot
    fig, ax = plt.subplots(1, figsize=(30, 4))
    ax.plot(x,y, color = plt_color)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)

    # configuring x axis ticks
    ax.set_xticks(ticks= x[::30], labels = x[::30], rotation=60) # extract x at intervals of 30

    return fig, ax

# Request the Temperature
x = time_series['time']
y = time_series['temperature_2m_mean']

fig, ax = lineplot(x,y,"Average Temperature in London", "Date", "Temperature [¬∞C]",plt_color = 'green')

#Request the Precipitation
y = time_series['precipitation_sum']
fig, ax = lineplot(x,y,"Daily Precipitation in London", "Date", "Precipitation [mm]",plt_color = 'red')





```

```python
def lineplots(x: ArrayLike, ys: List[ArrayLike], title: str, x_label: str, y_label: str, legend: List[str],plt_colors : list[str]) -> (plt.Figure, plt.Axes):
    """
    Creates a line plot for each list of data in the given list of data, title and labels.

    Args:
        x (ArrayLike): value of the data on the x-axis
        ys (List[ArrayLike]): list of values of the data on the y-axis
        title (str): title of the plot
        x_label (str): x-axis label
        y_label (str): y-axis label
        legend (List[str]): list of legend labels
        plt_color(list[str]): line color list of the plot
    Returns:
        (plt.Figure, plt.Axes): matplotlib figure and axes objects
    """

    

    # creating the plot
    fig, ax = plt.subplots(1, figsize=(30, 4))

    for i in range(len(ys)):
        ax.plot(x,ys[i], color = plt_colors[i])

    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)


    # configuring x axis ticks
    ax.set_xticks(ticks= x[::30], labels = x[::30], rotation=60)

    #Adding the legend
    ax.legend(legend)

    return fig,ax

#Request the plot
colors = ['green','red']
legend = ["Temperature [¬∞C]","Daily Precipitation [mm]"]
y = [time_series['temperature_2m_mean'],time_series['precipitation_sum']]
fig, ax = lineplots(x,y,"Temperature and Precipitation in London", "Date", "",legend,colors)
    
    
```

* The function should returns an array containing the normalised data
* Normalise the `temperature_2m_mean` feature

```python
import numpy as np


def minmax_norm(x: ArrayLike) -> ArrayLike:
    """
    Normalises the data between 0 and 1 using the min-max normalization.

    Args:
        x (ArrayLike): data to be normalized
    Returns:
        (ArrayLike): normalized data
    """
    # add your code here
    x = np.array(x)
    normalized_data = (x - x.min()) / (x.max() - x.min())

    return normalized_data


# get the data
x = time_series['time']
y = time_series['temperature_2m_mean']

#Resquest normalization
y_normalised = minmax_norm(y)
ys = [y_normalised,y]

#plot 
colors = ['blue','green']
legend = ["Normalized Temperature","Non-Normalized Temperature"]
fig, ax = lineplots(x,ys,"Normalized vs Not Normalized Temperature in London", "Date", "Temperature [¬∞C]",legend,colors)
```

<h4>üë©‚Äçüíªüë®‚Äçüíª [Optional] TASK 4.1: Z-score normalisation</h4>

* Create a function `zscore_norm(x)` that takes a pandas series and normalises it using the z-score.
* The function should returns an array containing the normalised data
* Normalise the `temperature_2m_mean` feature
* Plot the normalised data over the non-normalised one in the same plot

```python
def zscore_norm(x: ArrayLike) -> ArrayLike:
    """
    Normalises the data using the z-score normalization.

    Args:
        x (ArrayLike): data to be normalized
    Returns:
        (ArrayLike): normalized data
    """

    x = np.array(x)
    sigma = np.std(x)
    normalized_data = (x - x.mean())/sigma

    return normalized_data
    
# get the data
x = time_series['time']
y = time_series['temperature_2m_mean']
    
#Resquest normalization
y_normalised = zscore_norm(y)
ys = [y_normalised,y]

#plot 
colors = ['blue','green']
legend = ["Normalized Temperature","Non-Normalized Temperature"]
fig, ax = lineplots(x,ys,"Z Score Normalized vs Not Normalized Temperature in London", "Date", "Temperature [¬∞C]",legend,colors)
```

```python
def zscore_outliers(x: ArrayLike, threshold: int=3) -> ArrayLike:
    """
    Detects outliers using the z-score method and returns the indices of the outliers.

    Args:
        x (ArrayLike): data to be checked for outliers
        threshold (int): threshold for the z-score method

    Returns:
        (ArrayLike): indices of the outliers
    """
    #normalize data
    x_normalised = zscore_norm(x)

    #Extract indexes above threshold => outliers
    indexes =  np.where(x_normalised>threshold)

    return indexes
    

# Get outliers
y = time_series['temperature_2m_mean']
outlier_indexes = zscore_outliers(y,2)
```

* Create a function `iqr_detect(x, threshold=1.5)` that returns the indices of the outliers in the original data according to the **IQR** method

</div>

<div class="alert alert-heading alert-danger" style="background-color: white; border: 2px solid; border-radius: 5px; color: #000; border-color:#AAA; padding: 10px">
    <b>üíé Tip</b>

* You can use `np.quantile` to get the quantile of a uniform, discrete distrbution

</div>


```python
def iqr_detect(x: ArrayLike, threshold: float=1.5) -> ArrayLike:
    """
    Detects outliers using the interquantile range method and returns the indices of the outliers.

    Args:
        x (ArrayLike): data to be checked for outliers
        threshold (float): threshold for the interquantile range method

    Returns:
        (ArrayLike): indices of the outliers
    """

    # calculate the Q1 and Q3 quantiles
    x = np.array(x)
    Q1 = np.quantile(x, 0.25)
    Q3 = np.quantile(x, 0.75)

    #Whiskers min and max
    IQR = Q3-Q1
    IQR_min = Q1 - threshold*IQR
    IQR_max = Q3 + threshold*IQR

    #Extract outliers indexes
    indexes = np.where((x>IQR_max)|(x<IQR_min))

    return indexes

# Get outliers
y = time_series['temperature_2m_mean']
outlier_indexes = iqr_detect(y)


# Request the Temperature
x = time_series['time']
y = time_series['temperature_2m_mean']

#Plot outliers
fig, ax = lineplot(x,y,"Average Temperature in London", "Date", "Temperature [¬∞C]",plt_color = 'green')
ax = mark_outliers(ax,x,y,outlier_indexes)

```

<h4>üë©‚Äçüíªüë®‚Äçüíª TASK 6: Clamping outliers</h4>

* One way to deal with outliers is to compress them into a range of acceptable values
* Create a function `zscore_clamp(x, threshold=3)` that clamps the outliers into the range between 10th percentile and 90th percentile
* The function should return a numpy array containing the clamped data
* Plot the non-clamped data over the clamped one
  
</div>

<div class="alert alert-heading alert-danger" style="background-color: white; border: 2px solid; border-radius: 5px; color: #000; border-color:#AAA; padding: 10px">
    <b>üíé Tip</b>

* Not every outlier is bad!! For example in financial problems, such as stock prediction, outliers is **exactly** what you are looking for
* **Removing outliers in that case might not benefit our analysis**
* Always make an **informed** decision whether to remove outliers and always **justify** it

</div>


```python
def zscore_clamp(x: ArrayLike, threshold: int=3) -> ArrayLike:
    """
    Clamps outliers using the z-score method and returns the clamped data.

    Args:
        x (ArrayLike): data to be clamped
        threshold (int): threshold for the z-score method

    Returns:
        (ArrayLike): clamped data
    """

    #Detect outliers
    x = np.array(x)
    outlier_indexes = zscore_outliers(x,threshold)


    # Calculate percentile
    P_10 = np.quantile(x, 0.10)
    P_90 = np.quantile(x, 0.90)

    #Extract clamped data
    clamped_data = x.copy()

    for idx in outlier_indexes:
        clamped_data[idx] =  np.clip (x[idx], a_min=P_10, a_max=P_90)
    return clamped_data


# get the data
x = time_series['time']
y = time_series['temperature_2m_mean']

# clamp data
clamped_data = zscore_clamp(y,2)

ys = [y,clamped_data]

#Plot data
colors = ['red','green']
legend = ["Temperature","Clamped Temperature"]
fig, ax = lineplots(x,ys," Temperature in London", "Date", "Temperature [¬∞C]",legend,colors)



```

```python
def zscore_interp(x: ArrayLike, y: ArrayLike, threshold: int=3, deg: int=1) -> ArrayLike:
    """
    Interpolates outliers using the z-score method and returns the interpolated data using a `deg`-degree approximation.

    Args:
        x (ArrayLike): x-variate of data to be interpolated
        y (ArrayLike): y-variate of data to be interpolated
        threshold (int): threshold for the z-score method
        deg (int): degree of the polynomial approximation

    Returns:
        (ArrayLike): interpolated data
    """
    x = np.array(x)
    y = np.array(y)

    print(len(x))
    outlier_indexes = zscore_outliers(y,threshold)[0]
    interpolated_data = y.copy()

    #Clean the data
    x = np.arange(len(x)) # convert dates into integers, to use in the poly fit
    clean_time = np.delete(x,outlier_indexes)
    clean_data = np.delete(y,outlier_indexes)

    #Fit data in the model
    poly_fit = np.polynomial.Polynomial.fit(clean_time, clean_data, deg=deg)

    for idx in outlier_indexes:
        interpolated_data[idx] = poly_fit(x[idx])

    return interpolated_data


# get the data
x = time_series['time']
y = time_series['temperature_2m_mean']

# clamp data
interp_data = zscore_interp(x,y,threshold=1, deg=20)

ys = [y,interp_data]

#Plot data
colors = ['red','green']
legend = ["Temperature","interpolated Temperature"]
fig, ax = lineplots(x,ys," Temperature in London", "Date", "Temperature [¬∞C]",legend,colors)


    
```
